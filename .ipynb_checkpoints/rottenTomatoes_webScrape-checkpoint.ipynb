## Rotten Tomatoes Web Scraping Top 100 Movies Of 2021

```python
# load the packages
import requests
from bs4 import BeautifulSoup
```

```python
# Define the URL of the site
base_site = "https://www.rottentomatoes.com/top/bestofrt/?year=2021"
```

```python
# sending a request to the webpage
response = requests.get(base_site)
response.status_code
```

```python
# get the HTML from the webpage
html = response.content
```

```python
# convert the HTML to a BeatifulSoup object
soup = BeautifulSoup(html, 'lxml')
```

```python
soup
```

```python
# Find all div tags on the webpage containing the information we want to scrape
divs = soup.find_all("table", {"class": "table"})
```

```python
divs
```

```python
# Extracting all 'h2' tags
movies = [div.find_all("tr") for div in divs]
movies
```

```python
movies[0][0].find_all("th")
```
