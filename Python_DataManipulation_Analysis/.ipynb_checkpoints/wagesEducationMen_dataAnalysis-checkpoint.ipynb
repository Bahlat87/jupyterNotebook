# Looking At A Wages and Education of Young Males Dataset In Python With pandas & matplotlib

Hi there. In this Python project, I want to look at a dataset on wages and education of young males in Python with the help of the pandas, matplotlib and seaborn libraries. The dataset is from `https://catalog.data.gov/dataset/8th-grade-math-proficiency`. Information on the wages and education dataset is from [here](https://vincentarelbundock.github.io/Rdatasets/doc/Ecdat/Males.html).

```python
# Wages and Education of Young Males Dataset

'''
Website Source: https://vincentarelbundock.github.io/Rdatasets/csv/Ecdat/Males.csv

Download .csv file from the Download Link

Dataset Description:

""
a panel of 545 observations from 1980 to 1987

number of observations : 4360

observation : individuals

country : United States
""

school: years of schooling

wage: log of hourly wage

occupation: a factor with 9 levels

industry: a factor with 12 levels

Reference: Python For Data Analysis by Wes McKinney

'''

# Get current working directory: import os; os.getcwd()
# Set current working directory: import os; os.chdir(aPath)

import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Enable matplotlib in Jupyter Notebook
%matplotlib inline
```

## A Histogram Of (Log) Wages

With the use of matplotlib and seaborn, we can look at wages with a histogram.

```python
# Darkgrid style (looks like R's ggplot2):
    
sns.set_style("darkgrid")

plt.figure(figsize=(10, 8))# Resize plot

fig  = sns.distplot(wages_edu['wage'], kde = False)

plt.xlabel("\n (Log) Hourly Wages")
plt.ylabel("Count \n")
plt.title("A Histogram of Log Hourly Wages \n") 
plt.show(fig)
```

```python
# Hourly Wages (Reversing Natural Log)

# Import numpy as np:

import numpy as np

# Darkgrid style (looks like R's ggplot2):
    
sns.set_style("darkgrid")

plt.figure(figsize=(10, 8))# Resize plot

fig  = sns.distplot(np.exp(wages_edu['wage']), kde = False)

plt.xlabel("\n Hourly Wages")
plt.ylabel("Count \n")
plt.title("A Histogram of Hourly Wages \n") 
plt.show(fig)
```

**Some Summary Statistics On Wages**

With the use of Numpy functions, we can find some basic summary statistics on the wages in the dataset.


```python
# Summary Statistics

# Mean
wages_edu['wage'].mean()
```

```python
# Variance
wages_edu['wage'].var()
```

```python
# Standard deviation (square root of variance):
wages_edu['wage'].std()
```

```python
# Skewness

wages_edu['wage'].skew()
```

## Bar Graphs Of Occupation Counts

Bar graphs can help in visualization of occupation counts.


```python
# Reference: https://seaborn.pydata.org/generated/seaborn.barplot.html
# Reference: https://stackoverflow.com/questions/31632637/label-axes-on-seaborn-barplot

sns.set_style("whitegrid")

# Retrieve counts for each occupation:
# Reference: https://stackoverflow.com/questions/19384532/how-to-count-number-of-rows-per-group-and-other-statistics-in-pandas-group-by

wages_edu_byoccup = wages_edu.groupby(['occupation']).size().reset_index(name= 'Count')

# Print results
print(wages_edu_byoccup)
```

```python
plt.figure(figsize=(10, 8))# Resize plot

# Horizontal barplot
fig  = sns.barplot(y = "occupation", x = "Count", data= wages_edu_byoccup)

plt.xlabel("\n Count")
plt.ylabel("Occupation \n")
plt.title("Counts By Occupation In Dataset \n ", fontsize = 15) 
plt.show(fig)
```

## Bar Graphs Of Industry Counts 

Here is the code and output of indsutry counts of the young males.

```python
# Reference: https://seaborn.pydata.org/generated/seaborn.barplot.html
# Reference: https://stackoverflow.com/questions/31632637/label-axes-on-seaborn-barplot

sns.set_style("whitegrid")

# Retrieve counts for each occupation:
# Reference: https://stackoverflow.com/questions/19384532/how-to-count-number-of-rows-per-group-and-other-statistics-in-pandas-group-by

wages_edu_byindustry = wages_edu.groupby(['industry']).size().reset_index(name= 'Count')

# Print results
print(wages_edu_byindustry)
```

```python
plt.figure(figsize=(10, 8))# Resize plot

# Horizontal barplot
fig  = sns.barplot(y = "industry", x = "Count", data= wages_edu_byindustry)

plt.xlabel("\n Count")
plt.ylabel("Occupation \n")
plt.title("Counts By Occupation In Dataset \n ", fontsize = 15) 
plt.show(fig)
```

## Some Sample Scatterplots

In this section, some sample scatterplots are presented with the use of `matplotlib` and `seaborn`.

**Year Of Schooling vs Wage**

This first scatterplot will look at the years of schooling (x-axis) versus the wages (y-axis). Remember that wage is the log of the hourly wage.

```python
# Without regression fit:

plt.figure(figsize=(10, 8))# Resize plot
fig  = sns.regplot(x = wages_edu.school, y = wages_edu.wage, fit_reg=False)

plt.xlabel("\n Number Of Schooling Years")
plt.ylabel("(Log) Hourly Wages \n")
plt.title("Number Of Schooling Years Vs (Log) Wages \n") 
plt.show(fig)

sns.plt.show()
```

Recall the number of schooling years from the dataset are whole numbers. This is why we have straight lines of points on whole numbers of schooling years. It is difficult to see if there a trend as the number of schooling years increase. The next section is the same code as above except for `fit_reg = True`. Having `fit_reg = True` adds in a regression (trend) line.

```python
# With Regression fit:

plt.figure(figsize=(10, 8))# Resize plot
fig  = sns.regplot(x = wages_edu.school, y = wages_edu.wage, fit_reg= True)

plt.xlabel("\n Number Of Schooling Years")
plt.ylabel("(Log) Hourly Wages \n")
plt.title("Number Of Schooling Years Vs (Log) Wages \n") 
plt.show(fig)

sns.plt.show()
```

**Year Of Schooling vs Wage By Occupation**

To obtain a scatterplot by occupation, use the pairplot and the `hue = "occupation" ` option. 

```python
# Scatterplot Of Years Schooling Vs Wage By Occupation

plt.figure(figsize=(10, 8))# Resize plot
fig  = sns.pairplot(x_vars = ["school"], y_vars = ["wage"], hue = "occupation", data = wages_edu, size = 10);

plt.xlabel("\n Number Of Schooling Years")
plt.ylabel("(Log) Hourly Wages \n")
plt.title("Number Of Schooling Years Vs (Log) Wages By Occupation \n") 
plt.show(fig)

sns.plt.show()

```

**Year Of Schooling vs Wage By Industry**

To obtain a scatterplot by industry, use the pairplot and the `hue = "industry" ` option. 

```python
# Scatterplot Of Years Schooling Vs Wage By Industry

plt.figure(figsize=(10, 8))# Resize plot
fig  = sns.pairplot(x_vars = ["school"], y_vars = ["wage"], hue = "industry", data = wages_edu, size = 10);

plt.xlabel("\n Number Of Schooling Years")
plt.ylabel("(Log) Hourly Wages \n")
plt.title("Number Of Schooling Years Vs (Log) Wages By Industry\n") 
plt.show(fig)

sns.plt.show()
```
