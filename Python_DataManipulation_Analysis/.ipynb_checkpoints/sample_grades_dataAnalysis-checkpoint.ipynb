# A Simple Grades Dataset For Python Data Analysis

Hi there. I wanted to showcase a simple python data analysis project which involves a (made-up) grades dataset. There are about 30 students and each student has a score for two tests and a final exam.

```python
# Get current working directory: import os; os.getcwd()
# Set current working directory: import os; os.chdir(aPath)
# A good resource/reference is the seaborn documentation online.

import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Enable matplotlib in Jupyter Notebook
%matplotlib inline
```

### Loading In The Dataset 

```python
# Load in the grades dataset into Python/Jupyter Notebook:
# Use sep = "\t" to separate \t (tabs) in when reading .csv file.

grades_df = pd.read_csv("sample_grades01.txt", sep = "\t")


# Preview data with .head and .tail functions:
 
grades_df.head(15)
```

```python
# Preview data with .head and .tail functions:
 
grades_df.tail(15)
```

The `.describe` method can be used for quick summary statistics.

```python
grades_df.describe()
```

## Adding A New Column For Final Grades

In the dataset, we have two test scores and a final exam score for each student. We are not given a weighting scheme for the tests and the final score. This section will go over including an additional column for final grades with a weighting scheme.

```python
# Initalize Final Grade column as a new column

grades_df["Final Grade"] = 0

grades_df
```

```python
# Updating Final Grades scores with weighting scheme:
# Test One: 20 percent, Test Two: Percent and Final Exam as 50 percent

grades_df["Final Grade"] =  0.20 * grades_df["Test One"] + 0.30 * grades_df["Test Two"] + 0.5 * grades_df["Final Exam"]

# Alternatively, you can use .iloc with numeric indices

#grades_df["Final Grade"] =  0.20 * grades_df.iloc[:, 1] + 0.30 * grades_df.iloc[:, 2] + 0.5 * grades_df.iloc[:, 3]
```

```python
# Show the data with the final grades now:

grades_df
```

## Obtaining Some Statistics (Information) On The Data

The `.describe()` method is a quick and easy way to obtain some summary statistics but it is limited. What if we want to find out who got the highest (or lowest) score for each test and in the final exam. The `.max()` and `.min()` methods along with subsetting comes in handy.

```python
# Highest Final Grade

grades_df["Final Grade"].max()
```

```python
# Lowest Final Grade

grades_df["Final Grade"].min()
```

```python
# Average of Final Exam Scores 

grades_df["Final Exam"].mean()
```

```python
# Average of Final Grades 

grades_df["Final Grade"].mean()
```

```python
# Showing the student who got the highest final grade:

grades_df[grades_df["Final Grade"] == grades_df["Final Grade"].max()]
```

```python
# Showing the student who got the lowest final grade:

grades_df[grades_df["Final Grade"] == grades_df["Final Grade"].min()]
```

Sorting with a criteria is an option with the use of pandas. I sort the grades by highest final grade to the lowest final grade.

```python
# Sort students from highest final grade to lowest final grade.

sorted_grades = grades_df.sort_values(by = ["Final Grade"], ascending = False)

sorted_grades
```

```python
# Showing the student who got the second highest final grade (first row is index 0):

sorted_grades.iloc[1, :]
```

## Plotting Some Of The Grades Results

With the use of Python's matplotlib and seaborn, we can plot some histograms to give an idea how the test scores and the final grades are distributed.

```python
# Histogram of Test One Scores:
    
sns.set_style("darkgrid")

plt.figure(figsize=(10, 8))# Resize plot

fig  = sns.distplot(grades_df['Test One'], kde = False, bins = 20, color = "green")

plt.xlabel("\n Grade", size = 12)
plt.ylabel("Count \n", size = 12)
plt.title("Test One Scores \n", size = 20) 
plt.show(fig)
```

```python
# Histogram of Test Two Scores:
    
sns.set_style("darkgrid")

plt.figure(figsize=(10, 8))# Resize plot

fig  = sns.distplot(grades_df['Test Two'], kde = False, bins = 20, color = "red")

plt.xlabel("\n Grade", size = 12)
plt.ylabel("Count \n", size = 12)
plt.title("Test Two Scores \n", size = 20) 
plt.show(fig)
```

```python
# Histogram Of Final Exam Scores

sns.set_style("whitegrid")

plt.figure(figsize=(10, 8))# Resize plot

fig  = sns.distplot(grades_df['Final Exam'], kde = False, bins = 20, color = "blue")

plt.xlabel("\n Grade", size = 12)
plt.ylabel("Count \n", size = 12)
plt.title("Final Exam Scores \n", size = 20) 
plt.show(fig)
```

```python
# Histogram Of Final Grades

sns.set_style("whitegrid")

plt.figure(figsize=(12, 10))# Resize plot

fig  = sns.distplot(grades_df['Final Grade'], kde = False, bins = 20, color = "orange")

plt.xlabel("\n Grade", size = 12)
plt.ylabel("Count \n", size = 12)
plt.title("Final Grades \n", size = 20) 
plt.show(fig)
```

```python
# Multiple Plots In One Grid
# Reference: https://matplotlib.org/users/pyplot_tutorial.html

plt.figure(figsize=(20, 15))# Resize plot

# Test One
plt.subplot(221)
sns.distplot(grades_df['Test One'], kde = False, bins = 20, color = "green")
plt.xlabel("\n Grade", size = 12)
plt.ylabel("Count \n", size = 12)
plt.title("Test One Scores \n", size = 20) 


# Test Two
plt.subplot(222)
sns.distplot(grades_df['Test Two'], kde = False, bins = 20, color = "red")
plt.xlabel("\n Grade", size = 12)
plt.ylabel("Count \n", size = 12)
plt.title("Test Two Scores \n", size = 20) 

# Final Exam Scores

plt.subplot(223)
sns.distplot(grades_df['Final Exam'], kde = False, bins = 20, color = "blue")

plt.xlabel("\n Grade", size = 12)
plt.ylabel("Count \n", size = 12)
plt.title("Final Exam Scores \n", size = 20) 

# Final Grades

plt.subplot(224)
sns.distplot(grades_df['Final Grade'], kde = False, bins = 20, color = "orange")

plt.xlabel("\n Grade", size = 12)
plt.ylabel("Count \n", size = 12)
plt.title("Final Grades \n", size = 20) 


plt.show()
```

```python
# Another way to do histograms. With the use of pandas.
# https://pandas.pydata.org/pandas-docs/stable/visualization.html

grades_df.hist(color = 'red', alpha = 0.5, bins = 20, figsize=(15, 10))
```
