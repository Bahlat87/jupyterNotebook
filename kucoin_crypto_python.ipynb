## Creating A Function For mplfinance Plots

The above steps can be generalized into a function. It is assumed that the user puts in valid inputs and not crash the program.

```python
# Create function for plotting a crypto pair on Kucoin
# i.e BTC/USDT follows base/quote currency pair

# Time frame choices: 1hour, 1 day, 1week
# start_date

def plot_kucoin_crypto_pair(base_cur, quote_cur, time_frame, start_date, num_periods):
    
    # Trading pair:
    trade_pair = base_cur + '-' + quote_cur
    
    # Start Time (start_date as date(y, m, d)):
    unixtime = int(time.mktime(start_date.timetuple()))

    # End Time (today's date):
    today = datetime.today() 
    unixtime2 = int(time.mktime(today.timetuple()))
    
    
    # Obtain data 
    if time_frame == '1hour':
        user_data = client.get_kline_data(trade_pair, '1hour', unixtime, unixtime2)
    elif time_frame == '1day':
        user_data = client.get_kline_data(trade_pair, '1day', unixtime, unixtime2)
    elif time_frame == '1week':
        user_data = client.get_kline_data(trade_pair, '1week', unixtime, unixtime2)
    else:
        return print("Choose one of 1hour, 1day or 1week.")
    
    # Convert user_data into a pandas dataframe:
    df = pd.DataFrame(user_data)
    df.columns = ['Start Time', 'Open', 'Close', 'High', 'Low', 'Tsn Amount', 'Volume']
    
    # Convert into numeric data:
    df = df.apply(pd.to_numeric)
    
    # Start Time column into date format
    df['Start Time'] = pd.to_datetime(df['Start Time'] , unit = 's')
    
    # Sort in chronological order for start time:

    df =  df.sort_values(by='Start Time')

    #Plot with mplfinance with 10, 20 & 30 day Moving Averages
    
    mpf.plot(df.set_index("Start Time").tail(num_periods),
        type = 'candle', style = 'charles',
        title = trade_pair + ' Price \n', 
         ylabel = '\n' + base_cur + ' Price in ' + quote_cur,
        volume = True, mav = (10, 20, 30)
        )
```

```python
# Function Call:

d = date(2018,1,20)

plot_kucoin_crypto_pair("BTC", "USDT", time_frame = '1day', start_date = d, num_periods = 200)
```

```python
# Function Call Two ETH/USDT:

d = date(2018,1,20)

plot_kucoin_crypto_pair("ETH", "USDT", time_frame = '1day', start_date = d, num_periods = 200)
```
