## Adding In Currencies & Longer Time Period


We can add parameters for adding a desired base currency, other currencies and for obtaining exchanging exchange rates for longer time periods.

```python
# Base URL

url
```

```python
# Obtain CAD/USD Exchange Rates for the year of 2020 (Jan 1, 2020 start to Dec 31, 2020 end)

cadusd_2020_rates = "https://api.exchangeratesapi.io/history?start_at=2020-01-01&end_at=2020-12-31&symbols=USD&base=CAD"
cadusd_2020_rates
```

```python
# Extracting the response JSON object
data = requests.get(cadusd_2020_rates).json()
```

```python
# Data is not in chronological order

data
```

```python
# Check data type

type(data)
```

```python
data['rates']
```

```python
# Convert to DataFrame:

cadusd_df = pd.DataFrame.from_dict(data)

cadusd_df
```

```python
# Look at rates column:

cadusd_df['rates']
```

```python
# Extract the exchange rate value from USD:

rates_cadusd = cadusd_df['rates'].apply(lambda x: x['USD'])
rates_cadusd
```

```python
# Convert into pandas dataframe

rates_cadusd_df = pd.DataFrame(rates_cadusd)

rates_cadusd_df
```

```python
# Line Plot
# Not sure how to fix the xticks.

import seaborn as sns
import matplotlib.pyplot as plt


# Figsize
plt.figure(figsize=(14, 5))

plt.plot(rates_cadusd_df)
plt.xlabel("\n Date")
plt.ylabel("Exchange Rate (CAD/USD) \n")
plt.title("2020 Exchange Rates CAD/USD \n") 

plt.xticks(rotation ='vertical') 

plt.show()
```
